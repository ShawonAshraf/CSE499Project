```python
from Classifier.src.FruitClassifier.classifier import Classifier
```

```python
mango_images = [
    'Classifier/img_test/green_mangoes.jpg',
    'Classifier/img_test/rotten_mango.jpg',
    'Classifier/img_test/mango_001.jpg',
]

banana_images = [
    'Classifier/img_test/banana.jpeg'
]

images = mango_images + banana_images
```

```python
# specify label and graph path
label = 'Classifier/Training/retrained_labels.txt'
graph = 'Classifier/Training/retrained_graph.pb'
```

```python
# iterate for every image in the list

for image in images:
    cls = Classifier(image, label, graph)
    
    # view the image
    cls.plot_img()
    
    # now classify
    f, s = cls.classify()
    print('Fruit = {}\nScore = {}\n'.format(f, s))
    
print('\nDONE=================\n')
```
